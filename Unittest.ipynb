```python
# --- Application Code ---
class Calculator:
    def add(self, a, b):
        return a + b

    def divide(self, a, b):
        if b == 0:
            raise ValueError("Cannot divide by zero")
        return a / b

class CalculatorService:
    def __init__(self):
        self.calc = Calculator()

    def calculate_discount(self, total, percent):
        discount_amount = self.calc.divide(total * percent, 100)
        return total - discount_amount


# --- Integration Test ---
import unittest

class TestCalculatorService(unittest.TestCase):

    def test_discount(self):
        service = CalculatorService()
        result = service.calculate_discount(1000, 10)
        self.assertEqual(result, 900)

unittest.main(argv=[''], exit=False)
```

```python
# --- Application Code ---
class Calculator:
    def add(self, a, b):
        return a + b

class App:
    """Simulates user → system → result"""
    def __init__(self):
        self.calc = Calculator()

    def run(self, a, b):
        result = self.calc.add(a, b)
        return {"input": (a, b), "output": result}
```

```python
# --- End-to-End Test ---
import unittest

class TestEndToEnd(unittest.TestCase):

    def test_full_flow(self):
        app = App()

        # Simulating user input
        user_input_a = 10
        user_input_b = 20

        # Full system flow
        response = app.run(user_input_a, user_input_b)

        # Validate the complete process
        self.assertEqual(response["input"], (10, 20))
        self.assertEqual(response["output"], 30)

unittest.main(argv=[''], exit=False)
```

```python
class Calculator:

    def add(self, a, b):
        return a + b

    def subtract(self, a, b):
        return a - b

    def multiply(self, a, b):
        return a * b

    def divide(self, a, b):
        if b == 0:
            raise ValueError("Cannot divide by zero")
        return a / b
```

```python
import unittest

class TestCalculator(unittest.TestCase):

    def test_add(self):
        c = Calculator()
        self.assertEqual(c.add(10, 5), 15)

    def test_subtract(self):
        c = Calculator()
        self.assertEqual(c.subtract(10, 5), 5)

    def test_multiply(self):
        c = Calculator()
        self.assertEqual(c.multiply(10, 5), 50)

    def test_divide(self):
        c = Calculator()
        self.assertEqual(c.divide(20, 5), 4)

    def test_divide_by_zero(self):
        c = Calculator()
        with self.assertRaises(ValueError):
            c.divide(10, 0)


class CalculatorService:
    def __init__(self):
        self.calc = Calculator()

    def calculate_discount(self, total, percent):
        discount_amount = self.calc.divide(total * percent, 100)
        return total - discount_amount


class TestCalculatorService(unittest.TestCase):

    def test_discount(self):
        service = CalculatorService()
        result = service.calculate_discount(1000, 10)
        self.assertEqual(result, 900)


unittest.main(argv=[''], exit=False)
```
